```python
import os
import numpy as np

PES_file = ('Cl-CH3-OH_PES_scan.txt')
PES = np.genfromtxt(fname=PES_file, dtype='unicode')
headers = PES[0]
data = PES[1:]
data = data.astype(float)

coh = data[:,0]
clc = data[:,1]
energy = data[:,2]
```

```python
import matplotlib.pyplot as plt
from matplotlib import cm
# plot
fig, ax = plt.subplots()

ax.scatter(clc, coh, c=energy, vmin=min(energy), vmax=max(energy))

plt.show()
```

```python
# plot with 0 limit on energy value, make things more visible
fig, ax = plt.subplots()

ax.scatter(clc, coh, c=energy, vmin=min(energy), vmax=0)

plt.show()
```
